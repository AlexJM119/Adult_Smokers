```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.__version__
```

```python
data = pd.read_csv('share-of-adults-who-smoke.csv')
```

```python
albania = data[data.Entity == 'Albania']
cambodia = data[data.Entity == 'Cambodia']
haiti = data[data.Entity == 'Haiti']
```

```python
alb_year = albania['Year'].values
alb_tot = albania['Smoking prevalence, total (ages 15+)'].values
cam_year = cambodia['Year'].values
cam_tot = cambodia['Smoking prevalence, total (ages 15+)'].values
hat_year = haiti['Year'].values
hat_tot = haiti['Smoking prevalence, total (ages 15+)'].values
```

```python
smoke_data = pd.concat([albania, cambodia, haiti])
plt.plot(alb_year, alb_tot, label='Albania')
plt.plot(cam_year, cam_tot, label='Cambodia')
plt.plot(hat_year, hat_tot, label='Haiti')
plt.xlabel('Year')
plt.ylabel('Smoking prevalence, total (ages 15+)')
plt.title('Adult Smokers')
plt.legend()
```
